#### 기본 그래프 그리기

```python
import matplotlib.pyplot as plt
plt.plot([10,20,30,40])
plt.show()
```

```python
plt.plot([1,2,3,4], [10, 12, 14 ,20])
plt.show()
```

```python
#### 그래프에 범례 넣기
plt.title('legend')
plt.plot([10,20,30,40], label = 'asc')
plt.plot([40,30,20,10], label = 'desc')
plt.legend()
plt.show()
```

```python
#### 그래프 색상 추가
plt.title('color')
plt.plot([10,20,30,40], color = 'skyblue', label = 'asc')
plt.plot([40,30,20,10], color = 'hotpink', label = 'desc')
plt.legend() # 범례 표시
plt.show()
```

```python
# 그래프 선 모양 바꾸기
plt.title('linestyle') # 제목설정
plt.plot([10,20,30,40], color = 'skyblue', linestyle = '--' ,label = 'asc')
plt.plot([40,30,20,10], color = 'hotpink', ls=':' ,label = 'desc')
plt.legend() # 범례 표시
plt.show()
```

```python
plt.title('marker')
plt.plot([10,20,30,40], 'y*', label = 'star')
plt.plot([40,30,20,10], 'g^', label = 'triangle')
plt.legend() # 범례 표시
plt.show()
```

#### 기온 꺾은선 그래프

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
```

```python
next(data)
result = []
```

```python
for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))
```

```python
plt.plot(result,  color = 'hotpink')
```

```python
plt.figure(figsize=(10,2), dpi= 300)
plt.plot(result, color = 'hotpink')
plt.show()
```

```python
# 내생일의 최대 기온 값
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)
birth=[]
for row in data:
    month = row[0].split('-')[1]
    day = row[0].split('-')[2]
    if row[-1] != '':
        if month == "06" and day=='14' :
            birth.append(float(row[-1]))
```

```python
plt.figure(figsize=(10,2), dpi= 300)
plt.plot(birth, color = 'b')
plt.show()
```

```python
# 1월과 8월의 최대 기온 값

import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug=[]
jan=[]
for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == "08":
            aug.append(float(row[-1]))
        elif month == "01":
            jan.append(float(row[-1]))
        
```

```python
plt.hist(aug, bins=100, color="g")
plt.hist(jan,bins=100, color = 'y')
plt.show()
```
